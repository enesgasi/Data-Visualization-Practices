```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.plot([1,2,3],[2,4,1])
plt.grid()
plt.xlabel('X axis')
plt.ylabel('Y Axis')
plt.title('Graph')
```

```python
x=[10,20,30]
y=[40,10,30]
x1=[10,20,30]
y1=[20,40,10]
plt.plot(x1,y1,'b--',label='A-blue')
plt.plot(x,y,'r--',label='B-red',linewidth=8)
plt.grid()
plt.legend(loc=0)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plotting two different values')
plt.show()
```

```python
#f(x)=3*sin(x)-cos(x)
x = np.arange(0,4*np.pi,0.01)
y = np.sin(x)
z = np.cos(x)
f1 = 3*y-z

fig = plt.figure(figsize = (9,4));
ax = fig.add_axes([0,0,1,1])

ax.plot(x, f1, label='x')
ax.legend()
ax.grid()
ax.set_title('Multiple Graphs')
```

```python
#f(x)=3*sin(x)-cos(x)
#g(x)=2*cos(x)*sin(x)

x = np.arange(0,4*np.pi,0.01)
y = np.sin(x)
z = np.cos(x)
f1 = 3*y-z
f2 = 2*z*y


fig = plt.figure(figsize = (9,4))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, f1, label='f(x) = 3*sin(x)-cos(x)')
ax.plot(x, f2, label='g(x)= 2*cos(x)*sin(x)')
ax.legend()
ax.grid()
ax.set_title('Trigonometrical Graph')
```
